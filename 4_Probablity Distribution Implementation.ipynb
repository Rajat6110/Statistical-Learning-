# Probablity Distribution Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#NORMAL DISTRIBUTION >> GAUSSAIN DISTRIBUTION >> 
#MAJORITY OF DATA POINT IS CLUSTERD AROUND MEAN
```

```python
mu =  0 #mean
sigma = 0.1 #std dev

s =np.random.normal(mu , sigma , size = 10000)
```

```python
len(s)
```

```python
plt.hist(s, bins = 30)
```

```python
sns.distplot(s)
```

```python
sns.histplot(s , kde = True)
```

# log normal distribution 

```python
l = np.random.lognormal(mean = 0 , sigma = 0.5 ,size =  1000)
l = l[np.where(abs(l)< 5)]
```

```python
plt.hist(l , bins = 100)
```

```python
sns.boxplot(l) # since log normal is right skewed , outliers are present
```

```python
sns.distplot(l)
```

```python
#log normal becomes >> normal distribution
```

```python
sns.distplot(np.log(l))
```

```python
# sqaure root transformation
```

```python
sns.distplot(np.sqrt(l))
```

# chi Sqaure distribution 

```python
c = np.random.chisquare(df = 30 , size = 10000)
```

```python
c = c[np.where(abs(c)<100)]
plt.hist(c , bins = 1000)
```

```python
c = np.random.chisquare(df = 30, size = 10000)
c = c[np.where(abs(c) < 100)]
plt.hist(c, bins = 1000)
plt.show()
```

```python

c = np.random.chisquare(df = 30, size = 10000)
d = np.random.chisquare(df = 10, size = 10000)
c = c[np.where(abs(c) < 100)]
d = d[np.where(abs(d) < 100)]
plt.hist(c, bins = 1000)
plt.hist(d, bins = 1000)
plt.show()
```

```python
#multiple chi sq in one plot 
```

# T DISTRIBUTION 

```python
t = np.random.standard_t(df = 3 , size = 10000)
t = t[np.where(abs(t)< 100)]
plt.hist(t , bins = 1000)
```

```python
t = np.random.standard_t(df = 20 , size = 10000)
t = t[np.where(abs(t)< 100)]
plt.hist(t , bins = 1000)
```

# F distribution 

```python
f = np.random.f(100,100,10000)
f = f[np.where(abs(f)< 100)]
plt.hist(f , bins = 10000)
```

```python
f = np.random.f(dfnum= 100, dfden =100, size = 10000)
f = f[np.where(abs(f)< 100)]
plt.hist(f , bins = 10000)
```
